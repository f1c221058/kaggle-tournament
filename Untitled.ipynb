```python
# Import pustaka yang dibutuhkan
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd

# Set the path to the specific file you'd like to load from the dataset.
# GANTI "nama_file_anda.csv" DENGAN NAMA FILE YANG BENAR DARI DATASET INI.
file_path = "Crypto_historical_data.csv" # Contoh: Ganti dengan nama file sebenarnya, misalnya 'crypto_prices_data.csv'

# Load the latest version
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "isaaclopgu/cryptocurrency-historical-prices-top-100-2025",
    file_path,
    # Jika perlu, Anda dapat menambahkan pandas_kwargs di sini.
    # Contoh: pandas_kwargs={"sep": ";"}
)

print("First 5 records:", df.head())
```

# Cleaning Data

```python
# Mengubah kolom 'Date'
df['Date'] = pd.to_datetime(df['Date'], utc=True)
# Menghapus baris yang memiliki nilai NaN pada kolom harga
df_cleaned = df.dropna(subset=['Open', 'High', 'Low'])
# Mengisi nilai NaN (Not a Number) di masing-masing kolom dengan nilai mediannya
columns_to_impute = ['Open', 'High', 'Low']

for col in columns_to_impute:
    # Hitung median kolom saat ini
    median_value = df[col].median()
    
    # Isi nilai NaN dengan median
    df[col].fillna(median_value, inplace=True)

print("Pengisian nilai hilang (NaN) dengan median telah selesai.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Daftar Ticker yang ingin divisualisasikan
top_tickers = ['DOGE-USD', 'STETH-USD', 'TRX-USD', 'BNB-USD']

# Pastikan df sudah didefinisikan sebelumnya dan memiliki kolom 'ticker', 'Date', dan 'Close'
# Filter DataFrame hanya untuk ticker-ticker ini
df_top = df[df['ticker'].isin(top_tickers)]

# Buat grafik relplot dengan pembagian kolom (col)
g = sns.relplot(
    data=df_top, 
    x='Date', 
    y='Close', 
    kind='line', 
    col='ticker', 
    col_wrap=3, 
    height=3, 
    aspect=1.5,
    facet_kws={'sharey': False} # Setiap grafik memiliki sumbu Y-nya sendiri
)

# Sesuaikan judul
g.fig.suptitle('Grafik Harga Penutupan Berdasarkan Ticker', 
               y=1.02, # Posisikan judul di atas
               fontsize=16)

# Tampilkan plot
plt.show()
```
